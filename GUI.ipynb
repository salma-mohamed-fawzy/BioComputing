```python
#translation_test="GCCTCTGGTATGCTTCCCTGCTTGATGATTGCGGAAACCGTAACCCTACAAG"

#overlap_test: #a="ACGGTA" --  b="GTAGTA"

##SUFFIX-TEST= 'ACGACTACGATAAC$'

#Clustring  test
# x = 4, 5, 10, 4, 3, 11, 14 , 6, 10, 12
# y = 21, 19, 24, 17, 16, 25, 24, 22, 21, 21
# n_clusters=2

#good suufix rule 
#text = "CGTGCCTACTTACTTACTTACTTACGCGAA" 
#pattren = "CTTACTTAC"


#pip install customtkinter
import customtkinter
#import tkinter.font as font
import tkinter 
from tkinter import ttk
from matplotlib import style
style.use("ggplot")
from tkinter import * 
import numpy as np
import time
import cv2
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from PIL import Image


customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

root= customtkinter.CTk()
root.geometry('1200x700')
root.resizable(False, False)  #
root.title('Bio-Computing-Project')



def openNewWindow1():  #screen1
        customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
        customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

        root= customtkinter.CTk()
        root.geometry('1200x700')
        root.resizable(False, False)
        root.title('Bio-Computing-Project')
        frame= customtkinter.CTkFrame(master=root)
        frame.pack(pady=20,padx=60,fill='both',expand=True) 
        
        ###note the command  change with each function call
        
        customtkinter.CTkButton(master=frame, text="Complement" ,width=550, height=70,font=('arial',20), text_color="white" ,command=complement_main ).grid(column=0, row=0)
        customtkinter.CTkButton(master=frame, text="Reverse" ,width=550, height=70,font=('arial',20), text_color="white" ,command=reverse_main ).grid(column=1, row=0)
        customtkinter.CTkButton(master=frame, text="Reverse Complement" ,width=550, height=70,font=('arial',20), text_color="white" ,command=rev_comp ).grid(column=0, row=1)
        customtkinter.CTkButton(master=frame, text="GC-Content" ,width=550, height=70,font=('arial',20), text_color="white" ,command=GC_CONTENT_CALC ).grid(column=1, row=1)
        customtkinter.CTkButton(master=frame, text="Translation Table" ,width=550, height=70,font=('arial',20), text_color="white" ,command=translation ).grid(column=0, row=2)
        customtkinter.CTkButton(master=frame, text="Matching" ,width=550, height=70,font=('arial',20), text_color="white" ,command=matching).grid(column=1, row=2)
        customtkinter.CTkButton(master=frame, text="Bad Character Rule" ,width=550, height=70,font=('arial',20), text_color="white" ,command=bad_char_rule ).grid(column=0, row=3)
        customtkinter.CTkButton(master=frame, text="Index Sorting" ,width=550, height=70,font=('arial',20), text_color="white" ,command=index ).grid(column=1, row=3)
        customtkinter.CTkButton(master=frame, text="Overlap" ,width=550, height=70,font=('arial',20), text_color="white" ,command=overlaping).grid(column=0, row=4)
        customtkinter.CTkButton(master=frame, text="Suffix Construction" ,width=550, height=70,font=('arial',20), text_color="white" ,command=suffix).grid(column=1, row=4)
        customtkinter.CTkButton(master=frame, text="Concatmer Method" ,width=550, height=70,font=('arial',20), text_color="white" ,command=concat__mer ).grid(column=0, row=5)
        customtkinter.CTkButton(master=frame, text="Distance Betweeen two Seqence" ,width=550, height=70,font=('arial',18), text_color="white" ,command=Dis).grid(column=1, row=5)
        customtkinter.CTkButton(master=frame, text="Good Suffix Rule",  width=550, height=70,font=('arial',20), text_color="white" ,command=good_suffix_rule ).grid(column=0, row=6)
        customtkinter.CTkButton(master=frame, text="K-Mean Clustring", width=550, height=70,font=('arial',20), text_color="white" ,command=clustering).grid(column=1, row=6)
        customtkinter.CTkButton(master=frame, text="fibonacci Sequence",  width=550, height=70,font=('arial',20), text_color="white" ,command=FIBO_NACCI ).grid(column=0, row=7)
        customtkinter.CTkButton(master=frame, text="LCS",  width=550, height=70,font=('arial',20), text_color="white" ,command=LCS).grid(column=1, row=7)
        exit=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=400, height=60,font=('arial',20), text_color="white" ,command=root.destroy )#.grid(column=2, row=7)
        exit.place(relx=0.5, rely=0.9, anchor=tkinter.CENTER)
        root.mainloop()

def login():
    root.destroy()
    openNewWindow1()
def GC_CONTENT_CALC():
    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 

    label=customtkinter.CTkLabel(master=frame , text='DNA Sequence ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.085, rely=0.3, anchor=tkinter.CENTER)
    
    DNA_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    DNA_entry.pack(pady=240,padx=10)
    DNA_entry.place(x = 40, y = 280) 
    def GC_Content():
        dna_getter=DNA_entry.get()
        dna_getter1=dna_getter.upper()
        dna_getter2=list(str(dna_getter1))

        A_cnt=dna_getter2.count('A')
        C_cnt=dna_getter2.count('C')
        G_cnt=dna_getter2.count('G')
        T_cnt=dna_getter2.count('T')

        GC_percent=round(((C_cnt+G_cnt)/(A_cnt+C_cnt+G_cnt+T_cnt))*100)
        GC_toplt=GC_percent
        resulted_GC=("{0}%".format(GC_percent))

        GC_result_entry.insert(INSERT, resulted_GC)
        x_axis=["GC","AT"]

        datum=GC_toplt
        y_axis=[datum,(100-datum)]

        gc=plt.bar(x_axis,y_axis,label="GC vs AT content", color='b')
        gc=plt.legend(loc="best")
        gc=plt.title("GC content")
        gc=plt.savefig("GC_content.png")
        im2 = Image.open("GC_content.png")
        im2.show()

        
    label=customtkinter.CTkLabel(master=frame , text='GC content calculation ' , text_color="white",width=600,height=70,font=('Roboto bold',60) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)


    #DNA_entry=customtkinter.CTkEntry(master=frame ,placeholder_text=" Please enter sequence !"  ,width=500, height=60,placeholder_text_color='white',font=('arial',20)).place( relx=0.050,rely=0.2)
    #GC_result_entry=customtkinter.CTkEntry(master=frame ,width=500, height=60,placeholder_text_color='white',font=('arial',20)).place( relx=0.050,rely=0.6)


    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=GC_Content  ).place(relx=0.7,rely=0.3)
    
    label=customtkinter.CTkLabel(master=frame , text='result ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.050, rely=0.6, anchor=tkinter.CENTER)
    
    GC_result_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    GC_result_entry.pack(pady=250,padx=10)
    GC_result_entry.place(x = 40, y = 525) 

    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    #button.pack(pady=20,padx=10,anchor=tkinter.CENTER)
    root.title('GC content calculation')
    root.mainloop()
    
def rev_comp():
    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green
    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 
    
    label=customtkinter.CTkLabel(master=frame , text='DNA Sequence ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.085, rely=0.3, anchor=tkinter.CENTER)
    
    DNA_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    DNA_entry.pack(pady=240,padx=10)
    DNA_entry.place(x = 40, y = 280) 
    
    def Reverse_Complement():
        dna_getter=DNA_entry.get()
        dna_getter1=dna_getter.upper()
        s=list(dna_getter1)
        s=reversed(s)
        rev="".join(s)

        dic={"G":"C","C":"G","A":"T","T":"A"}
        c=list(rev)
        for i in range(len(rev)):
            c[i]=str(dic[rev[i]])
        c="".join(c)

        rev_com_entry.insert(INSERT, c)
        
    label=customtkinter.CTkLabel(master=frame , text='Reverse Complement calculation ' , text_color="white",width=600,height=70,font=('Roboto bold',60)  )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)

    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=Reverse_Complement  ).place(relx=0.7,rely=0.3)
    
    label=customtkinter.CTkLabel(master=frame , text='result ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.050, rely=0.6, anchor=tkinter.CENTER)
    
    rev_com_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    rev_com_entry.pack(pady=250,padx=10)
    rev_com_entry.place(x = 40, y = 525) 

    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('Reverese complement calculation')
    root.mainloop()
def bad_char_rule():
    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green
    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 
    
    label=customtkinter.CTkLabel(master=frame , text='DNA Sequence ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.085, rely=0.3, anchor=tkinter.CENTER)
    
    DNA_entry1=Entry(frame, fg='black', width=29,font=('arial',25) )
    DNA_entry1.pack(pady=240,padx=10)
    DNA_entry1.place(x = 40, y = 280) 

    label=customtkinter.CTkLabel(master=frame , text='DNA Sub-Sequence ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.095, rely=0.43, anchor=tkinter.CENTER)
    
    DNA_entry2=Entry(frame, fg='black',  width=29,font=('arial',25))
    DNA_entry2.pack(pady=240,padx=10)
    DNA_entry2.place(x = 40, y = 379) #subseq

    def Badchars():
        dna_getter=DNA_entry1.get()
        seq=dna_getter.upper()#seq

        dna_getter1=DNA_entry2.get()
        sub_seq=dna_getter1.upper() #subseq

        table=np.zeros([4,len(sub_seq)])     
        row=["C","G","A","T"]
        for i in range (4):
            num=-1
            for j in range (len(sub_seq)):
                if row[i]==sub_seq[j]:
                    table[i,j]=-1
                    num=-1
                else:
                    num+=1
                    table[i,j]=num
        x=-1
        i=0
        while(i<len(seq)-len(sub_seq)):
            if sub_seq==seq[i:i+len(sub_seq)]:
                x=i
            else:
                for j in range(i+len(sub_seq)-1,i-1,-1):
                    if seq[j] != sub_seq[int(j-i)]:
                        k=row.index(seq[j])
                        i+=table[k,j-i]
                        break
            i=int(i+1)
        badchar_entry.insert(INSERT, "Index = " + str(x))


    label=customtkinter.CTkLabel(master=frame , text='Bad Char Rule ' , text_color="white",width=600,height=70,font=('Roboto bold',60) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)

    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=Badchars  ).place(relx=0.7,rely=0.4)
    
    
    label=customtkinter.CTkLabel(master=frame , text='Result' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.075, rely=0.6, anchor=tkinter.CENTER)
    
    badchar_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    badchar_entry.pack(pady=250,padx=10)
    badchar_entry.place(x = 40, y = 525) 

    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('Bad Character Rule')
    root.mainloop()
    
def FIBO_NACCI():
    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green
    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 
     
        
    
    label=customtkinter.CTkLabel(master=frame , text='Enter Number' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.085, rely=0.3, anchor=tkinter.CENTER)
    
    entry1=Entry(frame, fg='black', width=29,font=('arial',25) )
    entry1.pack(pady=240,padx=10)
    entry1.place(x = 40, y = 280) 

    FibArray = [0, 1] 
    def fibonacci(n):

        # Check is n is less
        # than 0
        if n < 0:
            return("Incorrect input")

        # Check is n is less
        # than len(FibArray)
        elif n < len(FibArray):
            return FibArray[n]
        else:       
            FibArray.append(fibonacci(n - 1) + fibonacci(n - 2))
            return FibArray[n]

    def fib():
            entry_getter=int(entry1.get())
            x=fibonacci(entry_getter)
            fib_entry.insert(INSERT, x)



    label=customtkinter.CTkLabel(master=frame , text=' Fibonacci Calculation ' , text_color="white",width=600,height=70,font=('Roboto bold',60) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)

    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=fib  ).place(relx=0.7,rely=0.3)
    
    
    label=customtkinter.CTkLabel(master=frame , text='Fibonacci result ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.10, rely=0.6, anchor=tkinter.CENTER)
    
    fib_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    fib_entry.pack(pady=250,padx=10)
    fib_entry.place(x = 40, y = 525) 


    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('Fibonacci')
    root.mainloop()

def LCS() :
    
    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green
    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 

    
    label=customtkinter.CTkLabel(master=frame , text='First String' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.070, rely=0.3, anchor=tkinter.CENTER)
    
    entry1=Entry(frame, fg='black', width=29,font=('arial',25) )
    entry1.pack(pady=240,padx=10)
    entry1.place(x = 40, y = 280) 

    label=customtkinter.CTkLabel(master=frame , text='Second String' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.080, rely=0.43, anchor=tkinter.CENTER)
    
    entry2=Entry(frame, fg='black',  width=29,font=('arial',25))
    entry2.pack(pady=240,padx=10)
    entry2.place(x = 40, y = 379) 

    # Dynamic programming implementation of LCS problem

    # Returns length of LCS for X[0..m-1], Y[0..n-1]

    def lcs():
        X=entry1.get()
        Y=entry2.get()
        m=len(X)
        n=len(Y)

        L = [[0 for i in range(n+1)] for j in range(m+1)]

        # Following steps build L[m+1][n+1] in bottom up fashion. Note
        # that L[i][j] contains length of LCS of X[0..i-1] and Y[0..j-1]
        for i in range(m+1):
            for j in range(n+1):
                if i == 0 or j == 0:
                    L[i][j] = 0
                elif X[i-1] == Y[j-1]:
                    L[i][j] = L[i-1][j-1] + 1
                else:
                    L[i][j] = max(L[i-1][j], L[i][j-1])

            # Create a string variable to store the lcs string
        lcs = ""

        # Start from the right-most-bottom-most corner and
        # one by one store characters in lcs[]
        i = m
        j = n
        while i > 0 and j > 0:

            # If current character in X[] and Y are same, then
            # current character is part of LCS
            if X[i-1] == Y[j-1]:
                lcs += X[i-1]
                i -= 1
                j -= 1

            # If not same, then find the larger of two and
            # go in the direction of larger value
            elif L[i-1][j] > L[i][j-1]:
                i -= 1

            else:
                j -= 1

        # We traversed the table in reverse order
        # LCS is the reverse of what we got
        lcs = lcs[::-1]
        l_c_s="LCS of " + X + " and " + Y + " is " +"( " +lcs+ " )"
        lcs_entry.insert(INSERT, l_c_s) 




    label=customtkinter.CTkLabel(master=frame , text=' Longest Common Subsequence ' , text_color="white",width=600,height=70,font=('Roboto bold',60) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)

    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=lcs  ).place(relx=0.7,rely=0.3)
 
    label=customtkinter.CTkLabel(master=frame , text=' LCS result ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.065, rely=0.6, anchor=tkinter.CENTER)
    
    lcs_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    lcs_entry.pack(pady=250,padx=10)
    lcs_entry.place(x = 40, y = 525) 

    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('Longest Common Subsequence')
    root.mainloop()

def index():

    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green
    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 
    
    
    
    label=customtkinter.CTkLabel(master=frame , text='DNA Sequence ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.085, rely=0.3, anchor=tkinter.CENTER)
    
    entry1=Entry(frame, fg='black', width=29,font=('arial',25) )
    entry1.pack(pady=240,padx=10)
    entry1.place(x = 40, y = 280) 

    label=customtkinter.CTkLabel(master=frame , text=' number of elements in each index' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.14, rely=0.43, anchor=tkinter.CENTER)
    
    entry2=Entry(frame, fg='black',  width=29,font=('arial',25))
    entry2.pack(pady=240,padx=10)
    entry2.place(x = 40, y = 379) 

    def IndexSorted():
        se_q=entry1.get()
        seq=se_q.upper()
        l_n=entry2.get() #length of each of returned item
        ln=int(l_n)

        index = []
        for i in range(len(seq)-ln+1):
            index.append((seq[i:i+ln], str(i)))
        index.sort() 
        res = [' - '.join(i) for i in index]
        indexx = str(res)  ####
        Index_Sorted.insert(INSERT, indexx) 




    label=customtkinter.CTkLabel(master=frame , text=' Index Sorted ' , text_color="white",width=600,height=70,font=('Roboto bold',60) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)

    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=IndexSorted  ).place(relx=0.7,rely=0.4)

    
    label=customtkinter.CTkLabel(master=frame , text='result ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.050, rely=0.6, anchor=tkinter.CENTER)
    
    Index_Sorted=Entry(frame, fg='black', width=29,font=('arial',25) )
    Index_Sorted.pack(pady=250,padx=10)
    Index_Sorted.place(x = 40, y = 525) 
 
    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('Index Sorted')
    root.mainloop()

def Dis():

    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 

    def upload_file_1():
        global file1
        file1=tkinter.filedialog.askopenfilename()
        print(file1)

    def upload_file_2():
        global file2
        file2=tkinter.filedialog.askopenfilename()
        print(file2)

    label=customtkinter.CTkLabel(master=frame , text='First File' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.060, rely=0.3, anchor=tkinter.CENTER)

    ttk.Button(frame, text="Choose A File From Your PC", width=40,padding=10,command=lambda:upload_file_1()).place(x = 40, y = 280)

    #DNA_entry1=Entry(frame, fg='black', width=29,font=('arial',25) ).place(x = 40, y = 280) 
    #DNA_entry1.pack(pady=240,padx=10)
    #DNA_entry1.place(x = 40, y = 280) 

    label=customtkinter.CTkLabel(master=frame , text='Second File' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.070, rely=0.43, anchor=tkinter.CENTER)

    ttk.Button(frame, text="Choose A File From Your PC", width=40,padding=10,command=lambda:upload_file_2()).place(x = 40, y = 379) 

    #DNA_entry2=Entry(frame, fg='black',  width=29,font=('arial',25))
    #DNA_entry2.pack(pady=240,padx=10)
    #DNA_entry2.place(x = 40, y = 379) 
    def Distance():
        #file1=DNA_entry1.get()
        #file2=DNA_entry2.get()

        #file=open("dna1.fasta")
        l=[i for i in file1]
        seq=l[1][0:-1]
        seq="ATGTGTG"
        tag=3
        dic={}
        for i in range(0,len(seq)-tag):
            dic[seq[i:i+tag]]=dic.get(seq[i:i+tag],0)+1
        #file=open("dna2.fasta")
        l=[i for i in file2]
        seq2=l[1][0:]
        seq2="CATGTG"
        dic2={}
        for i in range(0,len(seq2)-tag):
            dic2[seq2[i:i+tag]]=dic2.get(seq2[i:i+tag],0)+1
        k=list(dic.keys())
        for i in range(len(k)):
            dic2[k[i]]=(dic2.get(k[i],0)-dic[k[i]])
        d=list(dic2.values())
        Sum=0
        for i in range(len(d)):
            Sum+=d[i]**2
        distance=Sum**(0.5)
        #print("distance = ",distance)
        dis_entry.insert(INSERT,"Distance = " + str(distance))



    label=customtkinter.CTkLabel(master=frame , text='Distance Betweeen two Seqences' , text_color="white",width=600,height=70,font=('Roboto bold',60) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)


    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=Distance  ).place(relx=0.7,rely=0.4)

    label=customtkinter.CTkLabel(master=frame , text='Ditance Result ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.080, rely=0.6, anchor=tkinter.CENTER)

    dis_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    dis_entry.pack(pady=250,padx=10)
    dis_entry.place(x = 40, y = 525) 

    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('Distance Betweeen two Seqences')
    root.mainloop()
    
def concat__mer():
    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 

    def upload_file():
        global file_1
        file_1=tkinter.filedialog.askopenfilename()
        print(file_1)

    label=customtkinter.CTkLabel(master=frame , text='Open File' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.066, rely=0.3, anchor=tkinter.CENTER)

    ttk.Button(frame, text="Choose A 'Fasta' File From Your PC", width=50,padding=20,command=lambda:upload_file()).place(x = 40, y = 280)

    label=customtkinter.CTkLabel(master=frame , text='K-mer or tag' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.08, rely=0.49, anchor=tkinter.CENTER)
    
    entry=Entry(frame, fg='black',  width=29,font=('arial',25))
    entry.pack(pady=240,padx=10)
    entry.place(x = 40, y = 430) 
    
    def concatmermain():
        file = open(file_1)
        l=[i for i in file]
        s=l[1][:-1]
        tag=int(entry.get())

        dict={}

        for i in range(0,len(s)-tag):
            dict[s[i:i+tag]]=dict.get(s[i:i+tag],0)+1

        head=['Tag_Seqence','Count']
        df=pd.DataFrame(dict.items(),columns=head)
        df.to_csv("concatmer_result.csv")

        label=customtkinter.CTkLabel(master=frame , text=' Rsult File is saved Successfully as " concatmer_result_file.csv " ' , text_color="white",font=('Roboto bold',25) )
        label.place(relx=0.35, rely=0.66, anchor=tkinter.CENTER)


    label=customtkinter.CTkLabel(master=frame , text='Concatmer Calculation' , text_color="white",width=600,height=70,font=('Roboto bold',60) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)


    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=concatmermain  ).place(relx=0.7,rely=0.31)
    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.48)
    root.title('Concatmer')
    root.mainloop()
    
###############################################################################################################
###############################################################################################################
###############################################################################################################
###############################################################################################################

def complement_main():
    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 

    label=customtkinter.CTkLabel(master=frame , text='DNA Sequence ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.085, rely=0.3, anchor=tkinter.CENTER)

    DNA_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    DNA_entry.pack(pady=240,padx=10)
    DNA_entry.place(x = 40, y = 280) 

    def Complement():
        dna_getter=DNA_entry.get()
        dna_getter0=dna_getter.upper()
        comp = ""
        for i in dna_getter0:
            if (i=="A"):
                comp= comp +"T"
            if (i=="T"):
                comp= comp +"A"
            if (i=="C"):
                comp= comp +"G"
            if (i=="G"):
                comp= comp +"C"
        cresult_entry.insert(tkinter.INSERT,comp)

    label=customtkinter.CTkLabel(master=frame , text='Complement calculation ' , text_color="white",width=600,height=70,font=('Roboto bold',60) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)
    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=Complement  ).place(relx=0.7,rely=0.3)

    label=customtkinter.CTkLabel(master=frame , text='Result ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.051, rely=0.6, anchor=tkinter.CENTER)

    cresult_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    cresult_entry.pack(pady=250,padx=10)
    cresult_entry.place(x = 40, y = 525) 

    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('Complement calculation')
    root.mainloop()    

def reverse_main():
    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 

    label=customtkinter.CTkLabel(master=frame , text='DNA Sequence ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.085, rely=0.3, anchor=tkinter.CENTER)

    DNA_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    DNA_entry.pack(pady=240,padx=10)
    DNA_entry.place(x = 40, y = 280) 

    def Reverse():
        dna_getter=DNA_entry.get()
        dna_getter0=dna_getter.upper()
        result_entry.insert(tkinter.INSERT, dna_getter0[::-1])

    label=customtkinter.CTkLabel(master=frame , text='Reverse calculation ' , text_color="white",width=600,height=70,font=('Roboto bold',60) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)
    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=Reverse  ).place(relx=0.7,rely=0.3)

    label=customtkinter.CTkLabel(master=frame , text='Result ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.051, rely=0.6, anchor=tkinter.CENTER)

    result_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    result_entry.pack(pady=250,padx=10)
    result_entry.place(x = 40, y = 525) 

    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('Reverse calculation')
    root.mainloop()    

def translation (): 
    
    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 

    label=customtkinter.CTkLabel(master=frame , text='DNA Sequence ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.085, rely=0.3, anchor=tkinter.CENTER)

    DNA_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    DNA_entry.pack(pady=240,padx=10)
    DNA_entry.place(x = 40, y = 280) 

    def Translation_Table():
        dna_getter=DNA_entry.get()
        dic = {"TTT" : "F", "CTT" : "L", "ATT" : "I", "GTT" : "V",
               "TTC" : "F", "CTC" : "L", "ATC" : "I", "GTC" : "V",
               "TTA" : "L", "CTA" : "L", "ATA" : "I", "GTA" : "V",
               "TTG" : "L", "CTG" : "L", "ATG" : "M", "GTG" : "V",
               "TCT" : "S", "CCT" : "P", "ACT" : "T", "GCT" : "A",
               "TCC" : "S", "CCC" : "P", "ACC" : "T", "GCC" : "A",
               "TCA" : "S", "CCA" : "P", "ACA" : "T", "GCA" : "A",
               "TCG" : "S", "CCG" : "P", "ACG" : "T", "GCG" : "A",
               "TAT" : "Y", "CAT" : "H", "AAT" : "N", "GAT" : "D",
               "TAC" : "Y", "CAC" : "H", "AAC" : "N", "GAC" : "D",
               "TAA" : "*", "CAA" : "Q", "AAA" : "K", "GAA" : "E",
               "TAG" : "*", "CAG" : "Q", "AAG" : "K", "GAG" : "E",
               "TGT" : "C", "CGT" : "R", "AGT" : "S", "GGT" : "G",
               "TGC" : "C", "CGC" : "R", "AGC" : "S", "GGC" : "G",
               "TGA" : "*", "CGA" : "R", "AGA" : "R", "GGA" : "G",
               "TGG" : "W", "CGG" : "R", "AGG" : "R", "GGG" : "G" 
               }
        s=""
        for i in range(0,len(dna_getter)-2,3):
            s+=dic[dna_getter[i:i+3]]
        result_entry1.insert(tkinter.INSERT, s)


    label=customtkinter.CTkLabel(master=frame , text='Translation Table' , text_color="white",width=600,height=70,font=('Roboto bold',60) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)
    button1=customtkinter.CTkButton(master=frame ,text="Translate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=Translation_Table  ).place(relx=0.7,rely=0.3)

    label=customtkinter.CTkLabel(master=frame , text='Amino Acids' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.070, rely=0.6, anchor=tkinter.CENTER)

    result_entry1=Entry(frame, fg='black', width=29,font=('arial',25) )
    result_entry1.pack(pady=250,padx=10)
    result_entry1.place(x = 40, y = 525) 

    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('Translation Table')
    root.mainloop() 
    
def matching():

    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 

    label=customtkinter.CTkLabel(master=frame , text='DNA Sequence ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.085, rely=0.3, anchor=tkinter.CENTER)

    DNA_entry1=Entry(frame, fg='black', width=29,font=('arial',25) )
    DNA_entry1.pack(pady=240,padx=10)
    DNA_entry1.place(x = 40, y = 280) 

    label=customtkinter.CTkLabel(master=frame , text='DNA Sub-Sequence ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.095, rely=0.43, anchor=tkinter.CENTER)

    DNA_entry2=Entry(frame, fg='black',  width=29,font=('arial',25))
    DNA_entry2.pack(pady=240,padx=10)
    DNA_entry2.place(x = 40, y = 379) #subseq
    def match():
        dna_getter=DNA_entry1.get()
        dna_getter0=dna_getter.upper()
        dna_getter1=DNA_entry2.get()
        dna_getter2=dna_getter1.upper()

        x=-1
        for i in range(len(dna_getter0)-len(dna_getter2)+1):
            if dna_getter2==dna_getter0[i:i+len(dna_getter2)]:
                x=i
        result_entry.insert(tkinter.INSERT, "Index = " + str (x))

    label=customtkinter.CTkLabel(master=frame , text='Matching Calculation' , text_color="white",width=600,height=70,font=('Roboto bold',60) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)
    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=match  ).place(relx=0.7,rely=0.3)

    label=customtkinter.CTkLabel(master=frame , text='Result' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.050, rely=0.6, anchor=tkinter.CENTER)

    result_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    result_entry.pack(pady=250,padx=10)
    result_entry.place(x = 40, y = 525)

    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('Matching')
    root.mainloop()  
    
def overlaping():
    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 

    label=customtkinter.CTkLabel(master=frame , text='Sequence 1 ' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.072, rely=0.3, anchor=tkinter.CENTER)

    DNA_entry1=Entry(frame, fg='black', width=29,font=('arial',25) )
    DNA_entry1.pack(pady=240,padx=10)
    DNA_entry1.place(x = 40, y = 280) 

    label=customtkinter.CTkLabel(master=frame , text='Sequence 2' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.070, rely=0.43, anchor=tkinter.CENTER)

    DNA_entry2=Entry(frame, fg='black',  width=29,font=('arial',25))
    DNA_entry2.pack(pady=240,padx=10)
    DNA_entry2.place(x = 40, y = 379) #subseq
    def Overlap():

        dna_getter1=DNA_entry1.get().upper()
        dna_getter2=DNA_entry2.get().upper()
        min_len=3
        j=len(dna_getter1)-1
        count=0
        for i in range(len(dna_getter2)-1,-1,-1):
            if dna_getter1[j]==dna_getter2[i]:
                count+=1
                j-=1
            else:
                count=0
                j=len(dna_getter1)-1
                if dna_getter1[j]==dna_getter2[i]:
                    count+=1
                    j-=1
        if count>=min_len:
            result_entry.insert(INSERT,"Index = " + str(count))
        else:
            result_entry.insert(INSERT, 'no overlap')

    label=customtkinter.CTkLabel(master=frame , text='Overlap Calculation' , text_color="white",width=600,height=70,font=('Roboto bold',60) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)
    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=Overlap  ).place(relx=0.7,rely=0.4)

    label=customtkinter.CTkLabel(master=frame , text='Result' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.050, rely=0.6, anchor=tkinter.CENTER)

    result_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    result_entry.pack(pady=250,padx=10)
    result_entry.place(x = 40, y = 525)

    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('Overlap')
    root.mainloop()    
    
def suffix():
    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 

    label=customtkinter.CTkLabel(master=frame , text='DNA Sequence' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.09, rely=0.3, anchor=tkinter.CENTER)
    
    DNA_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    DNA_entry.pack(pady=240,padx=10)
    DNA_entry.place(x = 40, y = 280) 
    def construct():
        dna_getter=DNA_entry.get()
        l=[]
        for i in range(len(dna_getter)):
            l.append(dna_getter[i:])
        l2=l.copy()
        l.sort()
        dec={}
        for i in range(len(l)):
            dec[l[i]]=i
        table=[]
        for i in range(len(l)):
            table.append([l2[i],i,dec[l2[i]]])
        result_entry.insert(INSERT, table)

    label=customtkinter.CTkLabel(master=frame , text='Constraction OF suffix by Character Method' , text_color="white",width=600,height=70,font=('Roboto bold',40) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)
    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=construct  ).place(relx=0.7,rely=0.3)

    label=customtkinter.CTkLabel(master=frame , text='Result' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.050, rely=0.6, anchor=tkinter.CENTER)

    result_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    result_entry.pack(pady=250,padx=10)
    result_entry.place(x = 40, y = 525)

    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('Constraction OF suffix by Character Method')
    root.mainloop()  
    
def clustering ():
    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 

    label=customtkinter.CTkLabel(master=frame , text='X - List' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.06, rely=0.3, anchor=tkinter.CENTER)

    entry1=Entry(frame, fg='black', width=29,font=('arial',25) )
    entry1.pack(pady=240,padx=10)
    entry1.place(x = 40, y = 280) 

    label=customtkinter.CTkLabel(master=frame , text='Y - List' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.06, rely=0.43, anchor=tkinter.CENTER)

    entry2=Entry(frame, fg='black',  width=29,font=('arial',25))
    entry2.pack(pady=240,padx=10)
    entry2.place(x = 40, y = 379) 

    label=customtkinter.CTkLabel(master=frame , text='Number of clusters' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.10, rely=0.55, anchor=tkinter.CENTER)

    entry3=Entry(frame, fg='black', width=29,font=('arial',25) )
    entry3.pack(pady=250,padx=10)
    entry3.place(x = 40, y = 500)

    def kmean():
        L1=entry1.get() #string
        L2=entry2.get()
        n_clusters=int(entry3.get() )

        x1=L1.split(",") #to convert to list
        y1=L2.split(",")

        ########
        result1=plt.scatter(x1, y1)
        result1=plt.title('The Data Repressentation Before clustring')
        result1=plt.savefig("Before_clustring.png")
        im1 = Image.open("Before_clustring.png")
        im1.show()
        ##########
        data = list(zip(x1, y1))
        kmeans = KMeans(n_clusters)
        kmeans.fit(data)
        ########
        
        result2=plt.scatter(x1, y1, c=kmeans.labels_)
        result2=plt.title('The Data Repressentation After clustring')
        result2=plt.savefig("After_clustring.png")
        im2 = Image.open("After_clustring.png")
        im2.show()


    label=customtkinter.CTkLabel(master=frame , text='K-Mean Clustering' , text_color="white",width=600,height=70,font=('Roboto bold',40) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)
    button1=customtkinter.CTkButton(master=frame ,text="cluster"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=kmean  ).place(relx=0.7,rely=0.57)

    #button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.59)
    root.title('kmean')
    root.mainloop()

def good_suffix_rule():

    customtkinter.set_appearance_mode("dark")    # Modes: system (default), light, dark
    customtkinter.set_default_color_theme('green')# Themes: blue (default), dark-blue, green

    root= customtkinter.CTk()
    root.geometry('1200x700')
    root.resizable(False, False)
    frame= customtkinter.CTkFrame(master=root)
    frame.pack(pady=20,padx=60,fill='both',expand=True) 

    label=customtkinter.CTkLabel(master=frame , text='DNA Text' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.06, rely=0.3, anchor=tkinter.CENTER)

    DNA_entry1=Entry(frame, fg='black', width=29,font=('arial',25) )
    DNA_entry1.pack(pady=240,padx=10)
    DNA_entry1.place(x = 40, y = 280) 

    label=customtkinter.CTkLabel(master=frame , text='DNA Pattern' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.07, rely=0.43, anchor=tkinter.CENTER)

    DNA_entry2=Entry(frame, fg='black',  width=29,font=('arial',25))
    DNA_entry2.pack(pady=240,padx=10)
    DNA_entry2.place(x = 40, y = 379) 

    def preprocess_strong_suffix(shift, bpos, pat, m):

        # m is the length of pattern
        i = m
        j = m + 1
        bpos[i] = j

        while i > 0:

            '''if character at position i-1 is 
            not equivalent to character at j-1, 
            then continue searching to right 
            of the pattern for border '''
            while j <= m and pat[i - 1] != pat[j - 1]:

                ''' the character preceding the occurrence 
                of t in pattern P is different than the 
                mismatching character in P, we stop skipping
                the occurrences and shift the pattern 
                from i to j '''
                if shift[j] == 0:
                    shift[j] = j - i

                # Update the position of next border
                j = bpos[j]

            ''' p[i-1] matched with p[j-1], border is found. 
            store the beginning position of border '''
            i -= 1
            j -= 1
            bpos[i] = j
    # Preprocessing for case 2
    def preprocess_case2(shift, bpos, pat, m):
        j = bpos[0]
        for i in range(m + 1):

            ''' set the border position of the first character 
            of the pattern to all indices in array shift
            having shift[i] = 0 '''
            if shift[i] == 0:
                shift[i] = j

            ''' suffix becomes shorter than bpos[0], 
            use the position of next widest border
            as value of j '''
            if i == j:
                j = bpos[j]

    #Search for a pattern in given text using 
    #Boyer Moore algorithm with Good suffix rule 
    def search():
        text=DNA_entry1.get()
        pat=DNA_entry2.get()
        # s is shift of the pattern with respect to text
        s = 0
        m = len(pat)
        n = len(text)

        bpos = [0] * (m + 1)

        # initialize all occurrence of shift to 0
        shift = [0] * (m + 1)

        # do preprocessing
        preprocess_strong_suffix(shift, bpos, pat, m)
        preprocess_case2(shift, bpos, pat, m)

        while s <= n - m:
            j = m - 1

            ''' Keep reducing index j of pattern while characters of 
                pattern and text are matching at this shift s'''
            while j >= 0 and pat[j] == text[s + j]:
                j -= 1

            ''' If the pattern is present at the current shift, 
                then index j will become -1 after the above loop '''
            if j < 0:
                result_entry.insert(INSERT, "pattern occurs at shift "+str(s)+"  ")
                result_entry.insert(INSERT, '\n')
                print("pattern occurs at shift = %d" % s)
                s += shift[0]
            else:

                '''pat[i] != pat[s+j] so shift the pattern 
                shift[j+1] times '''
                s += shift[j + 1]

    label=customtkinter.CTkLabel(master=frame , text='Good Suffix Rule' , text_color="white",width=600,height=70,font=('Roboto bold',40) )

    label.place(relx=0.5, rely=0.1, anchor=tkinter.CENTER)
    button1=customtkinter.CTkButton(master=frame ,text="Calculate"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=search  ).place(relx=0.7,rely=0.4)

    label=customtkinter.CTkLabel(master=frame , text='Result' , text_color="white",font=('Roboto bold',17) )
    label.place(relx=0.055, rely=0.6, anchor=tkinter.CENTER)

    result_entry=Entry(frame, fg='black', width=29,font=('arial',25) )
    result_entry.pack(pady=250,padx=10)
    result_entry.place(x = 40, y = 525) 

    button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=200, height=70,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.7,rely=0.6)
    root.title('good suffix rule')
    root.mainloop()    




    
frame= customtkinter.CTkFrame(master=root)
frame.pack(pady=20,padx=60,fill='both',expand=True) 

label=customtkinter.CTkLabel(master=frame , text=' Welcome To our Bio-Computing project ! ' , text_color="#fff", width=250,font=('Roboto bold',40) )
label.place(relx=0.5, rely=0.3, anchor=tkinter.CENTER)
#label.pack(pady=20,padx=10)


button1=customtkinter.CTkButton(master=frame ,text="Login"   ,width=400, height=80,font=('arial',30), text_color="white" ,command=login  ).place(relx=0.5, rely=0.5, anchor=tkinter.CENTER)
button2=customtkinter.CTkButton(master=frame ,text="Exit"   ,width=400, height=80,font=('arial',30), text_color="white" ,command=root.destroy  ).place(relx=0.5, rely=0.7, anchor=tkinter.CENTER)
#button.pack(pady=20,padx=10,anchor=tkinter.CENTER)

root.mainloop()

```
